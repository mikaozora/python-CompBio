```python
from Bio.Seq import Seq

SeqA = Seq("AGCTAGCTAGTGCAGCTGATCGAT")
print(SeqA.transcribe())
print(SeqA.translate())
print(SeqA.complement())
```

```python
from Bio.PDB.PDBParser import PDBParser
parser = PDBParser()
structure = parser.get_structure('6lu7', "6lu7.pdb")

model = structure[0]
for chain in model:
    print(chain)
    for residue in chain:
        for atom in residue:
            print(atom)
```

```python
#latian

seqC = Seq("ATGCGTTTATGACGTATCTGAAGTCTACTAGGGTGATCTGACTAA")
seqD = Seq("GTGTTAGTGCGCTAGTGCTGACTAGGTACAGTAGCTAGTTACTAG")
print(seqC.transcribe())
print(seqD.transcribe())

print()

seqE = seqC[0:24] + seqD[-21:]
seqE = Seq(str(seqE))
mRNAE = seqE.transcribe()

print("Seq A :", seqC.translate())
print("Seq B :",seqD.translate())
print("Seq C :", mRNAE)

#back_transcribe() untuk transkrip dari mRNA ke DNA
print("mRNA C ke DNA C:", mRNAE.back_transcribe())


print()
print("complement seq A: ", seqC.complement())
print("complement seq B: ", seqD.complement())
print("complement seq C: ", seqE.complement())
```
